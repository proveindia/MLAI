```python
import numpy as np
import pandas as pd
from scipy.linalg import svd
from sklearn.datasets import load_boston
X, _ = load_boston(return_X_y=True)
Xmean = np.mean(X, axis=0)
Xc = X - Xmean
U, s, Vt = svd(Xc,full_matrices=False)
Sigma = np.diag(s)
V = Vt.T
```

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(13)+1,s,linewidth=3,marker='o')
plt.grid()
```

```python
plt.plot(np.arange(13)+1,np.cumsum(s)/s.sum(),linewidth=3,marker='o')
plt.grid()
```
