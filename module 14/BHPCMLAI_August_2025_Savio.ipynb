```python
# IQR for outliers
import numpy as np

def find_boundaries(df, variable, distance):

  IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

  lower_bounds = df[variable].quantile(0.25) - (IQR * distance)
  upper_bounds = df[variable].quantile(0.75) + (IQR * distance)

  return lower_bounds, upper_bounds
```

```python
# the tukey formula = 1.5* IQR

lower_boundary, upper_boundary = find_boundaries(baseball, 'Salary', 0.7)
```

```python
lower_boundary, upper_boundary
```

```python
outliers = np.where(baseball['Salary'] > upper_boundary, True,
                    np.where(baseball['Salary'] < lower_boundary, True, False)
                    )
```

```python
# apply KMeans onto the principal components
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state = 0).fit(df_pca)
```

```python
kmeans.labels_
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_pca.iloc[:,0], df_pca.iloc[:,1], c = kmeans.labels_)
```

```python
sse = {}

for k in range(1,10):
  kmeans = KMeans(n_clusters=k, random_state = 0).fit(df_pca)
  sse[k] = kmeans.inertia_
  # inertia is the sum of the distances of the samples to theior closest center
```

```python
plt.plot(sse.keys(), sse.values())
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
```

```python
# % rate of change in SSE over the number of clusters
for a,b in zip(list(sse.values()), list(sse.values())[1:]):
  print((b-a)/a * 100)
```

```python
# optimal number of clusters is 5
kmeans = KMeans(n_clusters = 5, random_state = 0).fit(df_pca)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np

# Create example data
np.random.seed(0)
time = pd.date_range(start="2025-01-01", periods=200, freq="D")
trend = np.linspace(10, 20, 200)
season = 2 * np.sin(np.linspace(0, 12*np.pi, 200))
noise = np.random.normal(0, 0.5, 200)
data = trend + season + noise

series = pd.Series(data, index=time)

result = seasonal_decompose(series, model="additive")

result.plot()
plt.tight_layout()
plt.show()
```

```python
diff = series.diff().dropna()

plt.plot(diff)
plt.title("After First-Order Differencing (Stationary)")
plt.show()
```

```python
double_diff = diff.diff().dropna()

plt.plot(double_diff)
plt.title("After Second-Order Differencing (Stationary)")
plt.show()
```

```python
import itertools
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")  # ignore all warnings

# --- Train/test split (80/20) ---
y = data["Inbound_Calls"]
exog = data[["Marketing_Campaign", "Holiday"]]

# shuffle=False keeps chronological order
X_train, X_test, y_train, y_test = train_test_split(
    exog, y, test_size=0.2, shuffle=False
)

# --- Parameter ranges ---
p = q = range(0, 4)   # 0â€“3
d = [0, 1]
s = 7                 # weekly seasonality

pdqs = list(itertools.product(p, d, q))

best_mae = np.inf
best_order = None
best_seasonal_order = None
best_model = None

print("Searching models...")

for (p_, d_, q_) in pdqs:
    order = (p_, d_, q_)
    seasonal_order = (p_, d_, q_, s)   # same (p,d,q) but seasonal with s

    try:
        # Fit on training
        model = sm.tsa.statespace.SARIMAX(
            y_train,
            exog=X_train,
            order=order,
            seasonal_order=seasonal_order,
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        results = model.fit(disp=False)

        # Forecast on test
        forecast = results.get_forecast(steps=len(y_test), exog=X_test)
        y_pred = forecast.predicted_mean

        # Compute MAE
        mae = mean_absolute_error(y_test, y_pred)

        if mae < best_mae:
            best_mae = mae
            best_order = order
            best_seasonal_order = seasonal_order
            best_model = results
            print(f"New best model: ARIMA{seasonal_order} - MAE:{mae:.2f}")

    except Exception:
        continue

print("\nBest model found:")
print(best_model.summary())

```
